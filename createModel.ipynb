# Prepare data

```python
# addresses, tels, creditcards, emails, ssns

df = pd.DataFrame({'type':'address', 'value':addresses})
df = df.append(pd.DataFrame({'type':'telephone', 'value':tels}))
df = df.append(pd.DataFrame({'type':'creditcard', 'value':creditcards}))
df = df.append(pd.DataFrame({'type':'email', 'value':emails}))
df = df.append(pd.DataFrame({'type':'ssns', 'value':ssns}))
df = df.append(pd.DataFrame({'type':'website', 'value':websites}))
df = df.append(pd.DataFrame({'type':'city', 'value':cities}))
df = df.append(pd.DataFrame({'type':'country', 'value':countries}))
df = df.append(pd.DataFrame({'type':'other', 'value':other}))

df.describe()
```

# Create character map for LSTM

```python

raw_text = ''.join([x for x in list(df['value'].unique())])

# create mapping of unique chars to integers
chars = sorted(list(set(raw_text)))
char_to_int = dict((c, i) for i, c in enumerate(chars))
char_to_int
```

### Save sequence map

```python
import json
with open('charlookup.json', 'w') as f:
    f.write(json.dumps(char_to_int))
```

### Load sequence map and create sequence

```python
import numpy as np
import json
with open('charlookup.json', 'r') as f:
    char_to_int = json.loads(f.read())

max_length = 40
def generateSequence(input_string):
    output_seq = []
    for i in input_string[0:max_length]:
        output_seq.append(char_to_int[i])
    output_seq = np.array(output_seq)
    output_seq = np.pad(output_seq, (max_length - len(output_seq),0), 'constant', constant_values=(0, 0))
    return output_seq

print(generateSequence('Testing'))
```

```python
X = []
for i in df.iterrows():
    X.append(generateSequence(i[1]['value']))
    
X = np.array(X)
X = np.expand_dims(X, axis=1)
X
```

```python
categories = {value:key for (key, value) in enumerate(sorted(list(df['type'].unique())))}
lookups = {key:value for (key, value) in enumerate(sorted(list(df['type'].unique())))}
categories
```

```python
from keras.utils.np_utils import to_categorical   

Y = []
for i in df.iterrows():
    Y.append(categories[i[1]['type']])
    
Y = np.array(Y)
Y = to_categorical(Y, num_classes = len(categories))
Y
```

```python
from sklearn.model_selection import train_test_split
[X_train, X_test, Y_train, Y_test] = train_test_split(X, Y, test_size=0.15, random_state=42)
print(len(X_train))
print(len(X_test))
print(len(Y_train))
print(len(Y_test))
```

# Launch model

```python

import matplotlib.pyplot as plt

from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
```

```python
model = Sequential()
model.add(LSTM(max_length, input_shape=(1, max_length)))
model.add(Dropout(0.25))
model.add(Dense(len(categories), activation='softmax'))
model.compile(loss='categorical_crossentropy', optimizer='adadelta', metrics=['acc'])
model.fit(X_train, Y_train, epochs=200, batch_size=64)
```

```python
val = '38 Irving Ave, Ottawa, ON'
model.predict(np.expand_dims(np.expand_dims(np.array(generateSequence(val)), axis=0), axis=1))
```

```python
yhat = np.argmax(model.predict(X_test), axis=1)
ytest = np.argmax(Y_test, axis=1)
compsum = 0
for i in range(len(yhat)):
    if yhat[i] == ytest[i]:
        compsum +=1
        
print(compsum * 100 / len(yhat))
```

```python
from keras.models import model_from_json

model_json = model.to_json()
with open("datasorter.json", "w") as json_file:
    json_file.write(model_json)
# serialize weights to HDF5
model.save_weights("datasorter.h5")
print("Saved model to disk")
```
